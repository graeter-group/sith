```python
from vmol.view import VMolecule
from ase.io import read
```

```python
at = read('G.pdb')
```

```python
vm = VMolecule(at, alignment=[1,16, 9], default_bonds=True, background=[1,1,1])
color=vm._asvector([0, 0.631, 0.596])

t = vm.vp.label(text='1', pos=vm.vatoms[0].pos + vm._asvector([0, 1.3, 0]),
                color=color, align='center',
                opacity=0, box=False,
                height=30)
vm.vp.arrow(pos=vm.vatoms[0].pos + vm._asvector([0, 1, 0]),
            axis=vm._asvector([0, -0.75, 0]),
            color=color,
            round=True,
            shaftwidth=0.12)

vm.vp.arrow(pos=vm.vatoms[15].pos + vm._asvector([0, 1, 0]),
            axis=vm._asvector([0, -0.75, 0]),
            color=color,
            round=True,
            shaftwidth=0.12)

t = vm.vp.label(text='16', pos=vm.vatoms[15].pos + vm._asvector([0, 1.3, 0]),
                color=color, align='center',
                opacity=0, box=False, height=30)

#vm.vp.arro(pos=vm.vatoms[15].pos + vm._asvector([0.6, 2, 0]),
#            axis=vm._asvector([-0.25, -0.75, 0]),
#            color=vm._asvector([0.34,0.84,0.73]),
#            round=True)
```
